# Read the data, as you have already seen [here]( https://colab.research.google.com/drive/1Q3fNgpa8QfGTzJ7lj-oZ8Ya0IIGCdnEr).

```python
import pandas as pd
data = pd.read_excel('Data Set Feedcalculator.xlsx',sheet_name=None)
```

```python
ingredients      = data['Ingredient Database']
nutrient_rules   = data['Nutrient Rules']
ingredient_rules = data['Ingredient Rules']
```

# Collect the ingredients that are available together with the relevant columns

```python
available_ingredients = ingredients[ingredients.Availability][['Name','Reference name','Price']+list(nutrient_rules.Nutrient)].set_index('Reference name')
available_ingredients
```

# Collect the nutrient bounds

Note that 'not available' in a data frame is not the same as None!

```python
nutrient_bounds = { nut : (lb if not pd.isna(lb) else 0,     # no lower bound translates into 0 as lower bound
                           ub if not pd.isna(ub) else None)  # no upper bound becomes None
                    for nut,lb,ub in zip(nutrient_rules.Nutrient,nutrient_rules['Lower Bound'],nutrient_rules['Upper Bound'])
                  }
```

```python
nutrient_bounds
```

# Collect the ingredient bounds

Note again that 'not available' in a data frame is not the same as None!

```python
ingredient_bounds = { ing : (lb if not pd.isna(lb) else 0,    # no lower bound translates into 0 as lower bound
                             ub if not pd.isna(ub) else None) # no upper bound becomes None
                      for ing,lb,ub in zip(ingredient_rules.Ingredient,ingredient_rules['Lower Bound'],ingredient_rules['Upper Bound']) }
```

```python
ingredient_bounds
```

# Collect the combined ingredient rules

This is slightly more complex, as we need to know where this data is placed in the sheet, hence the data frame.
Note how we use `.dropna()` on series to leave only the values defined.
Note as well how we filter the ingredients that are not available.
It may happen that a combined rule disappears, because it did only relate to not available ingredients.

If you want to know where the `Unnamed: ` columns are comming from, just examine the `ingredient_rules` data frame.


```python
combined_ingredient_rules   = []
set_of_availabe_ingredients = set(available_ingredients.index)
for c in ['Unnamed: '+str(i) for i in range(5,13)]:
    aux = ingredient_rules[[c]].dropna().values
    aux = [ v[0] for v in aux ]
    upperbound          = aux[0]
    ingredients_in_rule = set(aux[1:]).intersection(set_of_availabe_ingredients)
    if ingredients_in_rule:
        combined_ingredient_rules.append((upperbound,ingredients_in_rule))
combined_ingredient_rules
```

# 1.1 Linear Model

```python
import pyomo.environ as pyo
def linear_model():
  model = pyo.ConcreteModel()

  model.I = pyo.Set(initialize = list(available_ingredients.index))
  def ing_bounds(model, i):
    if i in ingredient_bounds.keys():
      return(ingredient_bounds[i])
    else:
      return (0, None)
  model.N = pyo.Set(initialize = list(nutrient_rules.Nutrient))
  model.x = pyo.Var(model.I, within = pyo.NonNegativeReals, bounds = ing_bounds)

  model.objective = pyo.Objective(expr = pyo.quicksum(available_ingredients['Price'][i]*model.x[i] for i in model.I) , sense=pyo.minimize)

  def lower_bound_nutrients_in_mix_rule(model, N):
      return pyo.quicksum(available_ingredients[N][i]*model.x[i] for i in model.I) >= nutrient_bounds[N][0]
  model.lower_bound_nutrients_in_mix = pyo.Constraint(model.N, rule = lower_bound_nutrients_in_mix_rule)

  def upper_bound_nutrients_in_mix_rule(model, N):
      return pyo.quicksum(available_ingredients[N][i]*model.x[i] for i in model.I) <= nutrient_bounds[N][1]
  model.upper_bound_nutrients_in_mix = pyo.Constraint(model.N, rule = upper_bound_nutrients_in_mix_rule)

  def amount_of_mix_rule(model, I):
      return pyo.quicksum(model.x[i] for i in model.I) == 1
  model.amount_of_mix = pyo.Constraint(model.I, rule = amount_of_mix_rule)

  model.A = pyo.Constraint(expr = model.x["cotton"] + model.x["sunflower"] <= 0.17)
  model.B = pyo.Constraint(expr = model.x["soybeanexp"] + model.x["soybeanmeal"] <= 0.35)
  model.C = pyo.Constraint(expr = model.x["barley"] <= 0.3)
  model.D = pyo.Constraint(expr = model.x["maizebranhighq"] + model.x["maizebranlowq"] <= 0.25)
  model.E = pyo.Constraint(expr = model.x["fish"] + model.x["fishlq"] <= 0.07)
  model.F = pyo.Constraint(expr = model.x["mbmeal"] <= 0.08)
  model.G = pyo.Constraint(expr = model.x["casfine"] + model.x["caswhole"] + model.x["tapbran"] <= 0.25)

  return model

model = linear_model()
results = pyo.SolverFactory('cbc').solve(model)
print(results.solver.status, results.solver.termination_condition )
print( model.objective() )
[ x() for x in model.x.values() ]
```

# 1.2 Minimize the number of ingredients used

```python
import pyomo.environ as pyo
def minimize_the_number_of_ingredients_used():
  model = pyo.ConcreteModel()

  model.I = pyo.Set(initialize = list(available_ingredients.index))
  def ing_bounds(model, i):
    if i in ingredient_bounds.keys():
      return(ingredient_bounds[i])
    else:
      return (0, None)
  model.N = pyo.Set(initialize = list(nutrient_rules.Nutrient))
  model.x = pyo.Var(model.I, within = pyo.NonNegativeReals, bounds = ing_bounds)
  model.y = pyo.Var(model.I, within = pyo.Binary)

  model.objective = pyo.Objective(expr = pyo.quicksum(model.y[i] for i in model.I) , sense=pyo.minimize)

  def lower_bound_nutrients_in_mix_rule(model, N):
      return pyo.quicksum(available_ingredients[N][i]*model.x[i] for i in model.I) >= nutrient_bounds[N][0]
  model.lower_bound_nutrients_in_mix = pyo.Constraint(model.N, rule = lower_bound_nutrients_in_mix_rule)

  def upper_bound_nutrients_in_mix_rule(model, N):
      return pyo.quicksum(available_ingredients[N][i]*model.x[i] for i in model.I) <= nutrient_bounds[N][1]
  model.upper_bound_nutrients_in_mix = pyo.Constraint(model.N, rule = upper_bound_nutrients_in_mix_rule)

  def amount_of_mix_rule(model, I):
      return pyo.quicksum(model.x[i] for i in model.I) == 1
  model.amount_of_mix = pyo.Constraint(model.I, rule = amount_of_mix_rule)

  def binary_variable_rule(model, i):
    return model.y[i] >= model.x[i]
  model.binary_variable = pyo.Constraint(model.I, rule = binary_variable_rule)

  model.A = pyo.Constraint(expr = model.x["cotton"] + model.x["sunflower"] <= 0.17)
  model.B = pyo.Constraint(expr = model.x["soybeanexp"] + model.x["soybeanmeal"] <= 0.35)
  model.C = pyo.Constraint(expr = model.x["barley"] <= 0.3)
  model.D = pyo.Constraint(expr = model.x["maizebranhighq"] + model.x["maizebranlowq"] <= 0.25)
  model.E = pyo.Constraint(expr = model.x["fish"] + model.x["fishlq"] <= 0.07)
  model.F = pyo.Constraint(expr = model.x["mbmeal"] <= 0.08)
  model.G = pyo.Constraint(expr = model.x["casfine"] + model.x["caswhole"] + model.x["tapbran"] <= 0.25)

  return model

model = minimize_the_number_of_ingredients_used()
results = pyo.SolverFactory('cbc').solve(model)
print(results.solver.status, results.solver.termination_condition )
print( model.objective() )
[ x() for x in model.x.values() ]
```

# 1.3 Minimum cost for the minimum number of ingredients

```python
import pyomo.environ as pyo
def minimum_cost_for_the_minimum_number_of_ingredients(ingredient_limit):
  model = pyo.ConcreteModel()

  model.I = pyo.Set(initialize = list(available_ingredients.index))
  def ing_bounds(model, i):
    if i in ingredient_bounds.keys():
      return(ingredient_bounds[i])
    else:
      return (0, None)
  model.N = pyo.Set(initialize = list(nutrient_rules.Nutrient))
  model.x = pyo.Var(model.I, within = pyo.NonNegativeReals, bounds = ing_bounds)
  model.y = pyo.Var(model.I, within = pyo.Binary)

  model.objective = pyo.Objective(expr = pyo.quicksum(available_ingredients['Price'][i]*model.x[i] for i in model.I) , sense=pyo.minimize)

  def lower_bound_nutrients_in_mix_rule(model, N):
      return pyo.quicksum(available_ingredients[N][i]*model.x[i] for i in model.I) >= nutrient_bounds[N][0]
  model.lower_bound_nutrients_in_mix = pyo.Constraint(model.N, rule = lower_bound_nutrients_in_mix_rule)

  def upper_bound_nutrients_in_mix_rule(model, N):
      return pyo.quicksum(available_ingredients[N][i]*model.x[i] for i in model.I) <= nutrient_bounds[N][1]
  model.upper_bound_nutrients_in_mix = pyo.Constraint(model.N, rule = upper_bound_nutrients_in_mix_rule)

  def amount_of_mix_rule(model, I):
      return pyo.quicksum(model.x[i] for i in model.I) == 1
  model.amount_of_mix = pyo.Constraint(model.I, rule = amount_of_mix_rule)

  def binary_variable_rule(model, i):
    return model.y[i] >= model.x[i]
  model.binary_variable = pyo.Constraint(model.I, rule = binary_variable_rule)

  model.ytotal = pyo.Constraint(expr = pyo.quicksum(model.y[i] for i in model.I) <= ingredient_limit)

  model.A = pyo.Constraint(expr = model.x["cotton"] + model.x["sunflower"] <= 0.17)
  model.B = pyo.Constraint(expr = model.x["soybeanexp"] + model.x["soybeanmeal"] <= 0.35)
  model.C = pyo.Constraint(expr = model.x["barley"] <= 0.3)
  model.D = pyo.Constraint(expr = model.x["maizebranhighq"] + model.x["maizebranlowq"] <= 0.25)
  model.E = pyo.Constraint(expr = model.x["fish"] + model.x["fishlq"] <= 0.07)
  model.F = pyo.Constraint(expr = model.x["mbmeal"] <= 0.08)
  model.G = pyo.Constraint(expr = model.x["casfine"] + model.x["caswhole"] + model.x["tapbran"] <= 0.25)

  return model

model = minimum_cost_for_the_minimum_number_of_ingredients(5)
results = pyo.SolverFactory('cbc').solve(model)
print(results.solver.status, results.solver.termination_condition )
print( model.objective() )
[ x() for x in model.x.values() ]
```

# 1.4 Navigating the trade-off

```python
import matplotlib.pyplot as plt
x = list(range(5,14))# 5 because that's the minimum number of ingredients used and 14 because the optimal solution for the minimizing of the cost uses 13 ingredients
y = []
for i in x:
  model = minimum_cost_for_the_minimum_number_of_ingredients(i)
  results = pyo.SolverFactory('cbc').solve(model)
  y.append(model.objective())

plt.plot(x,y)
plt.xlabel('Number of Ingredients')
plt.ylabel('Cost')
plt.title('Feed Calculator Trade off')
plt.show()
```
